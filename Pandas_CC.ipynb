```python
#MAKING A T-TEST AND TAKING THE ONES WHERE THE NULL HYPOTHESIS IS REJECTED THE MOST STRONGLY

from scipy.stats import ttest_ind
films = list(set(data["title"].values))

def get_ttest(data, x):
    data = data[data["title"]==x]
    m = data[data["gender"]=="M"]["rating"].values
    f = data[data["gender"]=="F"]["rating"].values
    test = ttest_ind(m,f)
    return [x, test[1:2]]

ttest= [get_ttest(data, val) for val in films]
```

```python
test = pd.DataFrame(ttest, columns = ["title", "p-value"])
test.head()
```

10- Read data from csv file: `ma-ba.csv`. Count the number of times `Barça` wins `Madrid` and compute the stadistics of % win, % lose and % draw.

```python
data10 = pd.read_csv("ma-ba-utf8.csv")
data10.columns = ["ID", "Fecha", "(Lugar)", "Partidos", "resultado", "Competicion"]

data10["Partidos"] = data10["Partidos"].str.replace("Madrid FC  ", "Madrid FC ")
data10["Partidos"] = data10["Partidos"].str.replace("Madrid FC", "Madrid FC ")
data10["Partidos"] = data10["Partidos"].str.replace("R. Madrid", "Madrid FC ")
data10["Partidos"] = data10["Partidos"].str.replace("R.Madrid ", "Madrid FC ")

data10[['home team','away team']] = data10['Partidos'].str.split(' - ',expand=True)
data10[['home goals','away goals']] = data10['resultado'].str.split('-',expand=True)


def winner(df):
    if df["home goals"] > df["away goals"]:
        val = df["home team"]
    elif df["away goals"] > df["home goals"]:
        val = df["away team"]
    else:
        val = "draw"
    return val

data10["Winner"] = data10.apply(winner, axis =1)



b_win = (len(data10[data10["Winner"]=="Barcelona"])/len(data10))*100
r_win = (len(data10[data10["Winner"]=="Madrid FC "])/len(data10))*100
draws = (len(data10[data10["Winner"]=="draw"])/len(data10))*100
print(f' Barcelona won {b_win:.2f}% of matches')
print(f' Real Madrid won {r_win:.2f}% of matches')
print(f' Both teams drew {draws:.2f}% of matches')
```
